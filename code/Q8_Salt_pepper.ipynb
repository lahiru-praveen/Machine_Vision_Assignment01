```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img_path = '../images/emma_salt_pepper.jpg' 
img = cv2.imread(img_path)

if img is None:
    print("Error: Load image first.")
else:
    # Apply Gaussian Smoothing
    gaussian_blur = cv2.GaussianBlur(img, (5, 5), 0)

    # Apply Median Filtering
    median_blur = cv2.medianBlur(img, 5)

    # Save Outputs
    output_dir = '../images/outputs'
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    cv2.imwrite(os.path.join(output_dir, 'q8_gaussian.jpg'), gaussian_blur)
    cv2.imwrite(os.path.join(output_dir, 'q8_median.jpg'), median_blur)
    print("Saved Q8 results.")

    # Visualization
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gaussian_rgb = cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB)
    median_rgb = cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(15, 6))

    # Original Noisy Image
    plt.subplot(1, 3, 1)
    plt.imshow(img_rgb)
    plt.title('Original (Salt & Pepper Noise)')
    plt.axis('off')

    # Gaussian Result
    plt.subplot(1, 3, 2)
    plt.imshow(gaussian_rgb)
    plt.title('Gaussian Blur (Smudged Noise)')
    plt.axis('off')

    # Median Result
    plt.subplot(1, 3, 3)
    plt.imshow(median_rgb)
    plt.title('Median Filter (Clean Result)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
