```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# 1. Load Image
img_path = '../images/runway.png' 
img = cv2.imread(img_path) # Load in Color (BGR) to see the effect nicely

if img is None:
    print("Error: Load image first.")
else:
    # Define a Sharpening Kernel
    kernel_sharpen = np.array([[0, -1, 0],
                               [-1, 5, -1],
                               [0, -1, 0]])

    # Apply the Kernel
    sharpened = cv2.filter2D(img, -1, kernel_sharpen)

    # Save Output
    output_dir = '../images/outputs'
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        
    cv2.imwrite(os.path.join(output_dir, 'q9_runway_sharpened.jpg'), sharpened)
    print(f"Saved Q9 result to {output_dir}/")

    # Visualization
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    sharpened_rgb = cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(15, 8))
    
    # Original
    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title('Original Image')
    plt.axis('off')
    
    # Sharpened
    plt.subplot(1, 2, 2)
    plt.imshow(sharpened_rgb)
    plt.title('Sharpened Image')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```
